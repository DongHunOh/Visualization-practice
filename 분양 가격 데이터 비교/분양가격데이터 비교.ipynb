# 1.8.2 연도와 월을 분리하기

```python
date = "2013년12월"
date
```

```python
# split을 통해 "년"을 기준으로 텍스트를 분리해 봅니다.
# 원래 split은 공백을 기준으로 나눠지게 됨
# 나눠진 부분들은 list형태로 저장
date.split("년")
```

```python
date.split("년")[0]
```

```python
date.split("년")[1]
```

```python
# -1하게되면 마지막꺼 불러옴
date.split("년")[-1]
```

```python
date.split("년")[-1].replace("월", "")
```

```python
# parse_year라는 함수 만들고
# 연도만 반환하도록 하며, 반환하는 데이터는 int 타입이 되도록 한다
def parse_year(date):
    year = date.split("년")[0]  
    year = int(year)                  # int형으로 변환
    return year

y = parse_year(date)
print(type(y))
y
```

```python
# 제대로 분리 되었는지 parse_year 함수 확인
parse_year(date)
```

```python
# parse_month 라는 함수 만들고
# 월만 반환하도록 하며, 반환하는 데이터는 int타입이 되도록
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)   # 한글과 숫자 같이있을때 int로 변환 불가
    return month

```

```python
parse_month(date)
```

```python
df_first_melt
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서
# apply를 활용해 연도만 추출해서 새로운 컬럼에 담는다
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서
# apply를 활용해 월만 추출해서 새로운 컬럼에 담는다
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt
```

# 1.8.4 pivot_table 사용하기

```python
# 연도를 인덱스로, 지역명을 컬럼으로 평당분양가격을 피봇테이블로 그려보자
t = pd.pivot_table(df, index = "연도", columns="지역명", values="평당분양가격").round()
# 숫자라 한눈에 안들어옴
# 그래서 t라고 하고
t
```
